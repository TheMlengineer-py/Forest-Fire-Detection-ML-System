# Evaluating if dataset is balanced or not

```python
#Evaluating if dataset is balanced or not
percentage_fire = (Forest_Fire['Classes'].value_counts()['fire'] / (Forest_Fire['Classes'].value_counts()['fire'] + Forest_Fire['Classes'].value_counts()['not fire']))*100
percentage_notfire = (Forest_Fire['Classes'].value_counts()['not fire'] / (Forest_Fire['Classes'].value_counts()['fire'] + Forest_Fire['Classes'].value_counts()['not fire']))*100

print(f"Percentage of fire according to dataset is {round(percentage_fire)}%.\nand")
print(f"Percentage of non-fire according to dataset is {round(percentage_notfire)}%.")

#Threshold value to evaluate the difference between the dataset will be greator than the Threshold value
Threshold = 1 
#zero difference is acceptable

if(abs(percentage_fire- percentage_notfire) > Threshold):
    print("Imbalanced dataset")
else:
    print("Balanced dataset")
```

# Synthetic Minority Oversampling Technique or SMOTE is another technique to oversample the minority class

```python
from imblearn.over_sampling import SMOTE
# Resampling the minority class. The strategy can be changed as required.
sm = SMOTE(sampling_strategy='minority', random_state=42)
# Fit the model to generate the data.
oversampled_X, oversampled_Y = sm.fit_resample(Forest_Fire.drop('Classes', axis=1), Forest_Fire['Classes'])
oversampled = pd.concat([pd.DataFrame(oversampled_Y), pd.DataFrame(oversampled_X)], axis=1)
```

```python
#creating feature list
#creating IQR list
feature_list = integer_columns+decimal_columns
IQR = []
for feature in feature_list:
    if feature != "Classes":
        IQR.append(oversampled[feature].quantile(0.75) - oversampled[feature].quantile(0.25))
        
IQR_dict = dict(zip(feature_list, IQR)) #creating dictionary
print(IQR_dict) #Inter Quartile Range method was used in fixng the right skewness/higher outlier from the features in the above boxplots
```

```python
#getting the upper_bridge values
extreme_outliers = ["Ws", "Temperature", "BUI", "ISI", "DC", "DMC", "FFMC", "Rain"]
upper_bridge_val = []

for feature in extreme_outliers:
    upper_bridge_val.append(oversampled[feature].quantile(0.75) + (IQR_dict[feature]*1.5))
print(upper_bridge_val)

upper_bridge = dict(zip(extreme_outliers,upper_bridge_val))
print(upper_bridge)
```

```python
oversampled_no_outliers = oversampled.copy()
for feature in extreme_outliers:
    oversampled_no_outliers.loc[oversampled_no_outliers[feature] > upper_bridge[feature], feature] = upper_bridge[feature]
    
```

# Feature Selection using SelectKBest METHOD

```python
#loading dataset
oversampled
```

```python
#importing libraries
from sklearn.feature_selection import SelectKBest, chi2, f_regression, f_classif

#specifying targeted coloumns 
predictors = ["month", "Temperature", "RH", "Ws", "Rain", "FFMC", "DMC", "DC", "ISI", "BUI", "FWI"]

#drop negative values
usethis = oversampled.dropna()
usethis.reset_index(drop=True)

usethis = usethis.filter(["month", "Temperature", "RH", "Ws", "Rain", "FFMC", "DMC", "DC", "ISI", "BUI", "FWI", "Classes"], axis=1)

usethisnow = usethis[(usethis["month"]>0) & (usethis["Temperature"]>0) & (usethis["RH"]>0)
                     & (usethis["Ws"]>0) & (usethis["Rain"]>0) & (usethis["FFMC"]>0) 
                     & (usethis["DMC"]>0) & (usethis["DC"]>0) & (usethis["ISI"]>0) & (usethis["BUI"]>0) 
                     & (usethis["FWI"]>0)]
usethisnow
#top N conditions are for target variable (i.e., Classes)
# Perform the feature selection
selector = SelectKBest(f_classif, k="all")
selector.fit(usethisnow[predictors], usethisnow["Classes"])

#get the p-values for each feature and then transform from p-values into some scores
scores = -np.log10(selector.pvalues_)

# Using plt to plot the scores!

plt.bar(range(len(predictors)), scores)
plt.xticks(range(len(predictors)), predictors, rotation='vertical')
plt.show()
```

```python
#Investigating further which sets of feature fit the clasifiaction task 
#specifying targeted coloumns 
#The classification of the output classes of the dataset was achieved with the use of two parameters of the FWI (fire weather index) 
#i.e., FFMC (Fine Fuel Moisture Code) which indicates an ignition probability and derived from four weather features using the Fire weather index calculator.
#weather fEATURES; Temperature, Windspeed, Relative humidity, Rainfall

predictors = ["Temperature", "RH", "Ws", "Rain"]

#drop negative values
usethis = oversampled.dropna()
usethis.reset_index(drop=True)

usethis = usethis.filter(["Temperature", "RH", "Ws", "Rain", "Classes"], axis=1)

usethisnow = usethis[(usethis["Temperature"]>0) & (usethis["RH"]>0)
                     & (usethis["Ws"]>0) & (usethis["Rain"]>0)]

                    
usethisnow
#top N conditions are for target variable (i.e., Classes)
# Perform the feature selection
selector = SelectKBest(f_classif, k="all")
selector.fit(usethisnow[predictors], usethisnow["Classes"])

#get the p-values for each feature and then transform from p-values into some scores
scores = -np.log10(selector.pvalues_)

# Using plt to plot the scores!

plt.bar(range(len(predictors)), scores)
plt.xticks(range(len(predictors)), predictors, rotation='vertical')
plt.show()
```

```python
#use a hypothesis test to see if the observed results are statistically significant.
#use the 5×2 procedure to evaluate the algorithms and calculate a p-value and test statistic value
from mlxtend.evaluate import paired_ttest_5x2cv
# check if difference between algorithms(Support vector machine and random forest classifier) is real
t, p = paired_ttest_5x2cv(estimator1=svc, estimator2=rfc, X=X, y=y, scoring='accuracy', random_seed=1)
# summarize
print('P-value: %.3f, t-Statistic: %.3f' % (p, t))
```

```python
#interpret the p-value using an alpha of 5 percent.

# interpret the result
if p <= 0.05:
    print('Difference between mean performance is probably real')
else:
    print('Algorithms probably have the same performance')
    
#In this case, we can see that the p-value is about 0.717, which is much larger than 0.05. This leads us to fail to reject the null hypothesis, suggesting that any observed difference between the algorithms is probably not real.

#We could just as easily choose support vector machine or Random forest classifier and both would perform about the same on average.
```

# Model Training with test-train split

```python
#load dataset
#load dataset
oversampled

features = ["Temperature", "RH", "Ws", "Rain"]
target = ['Classes']

#drop negative values
usethis = oversampled.dropna()
usethis.reset_index(drop=True)

usethis = usethis.filter(["Temperature", "RH", "Ws", "Rain", "Classes"], axis=1)

usethisnow = usethis[(usethis["Temperature"]>0) & (usethis["RH"]>0)
                     & (usethis["Ws"]>0) & (usethis["Rain"]>0)]


from sklearn.model_selection import train_test_split
y_target = usethisnow["Classes"].values
Xfeatures = usethisnow.drop("Classes", axis=1)

#The new bit is to use test-train split like so: 80:20

X_train, X_test, y_train, y_test = train_test_split(Xfeatures,y_target,test_size=0.2,random_state=1)    
```

# Feature Scaling 

```python
#Feature Scaling of data
#oversampled dataset may contain features highly varying in magnitudes, range, and units. We need to bring all features to the same level of magnitudes. This can be done by Scaling the data, which means fitting the data within a specific range(example: 0-1).
#Using StandardScaler method in Scikit-Learn Library for scaling our data.
#importing library
from sklearn.preprocessing import StandardScaler 
ss = StandardScaler()
X_train = ss.fit_transform (X_train)
X_test = ss.fit_transform (X_test)
```
